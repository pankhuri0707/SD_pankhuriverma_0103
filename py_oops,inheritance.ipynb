```python
class students:
  def __init__(self,id,name,marks):
    self.id=id
    self.name=name
    self.marks=marks
  def display(self):
    print("id",self.id)
    print("name",self.name)
    print("marks",self.marks)
s1=students(100,"abc",90)
s1.display()
```

```python
class parents:
  def t_parent(self):
    print("this is my parent class")
```

```python
class child(parents):
  pass
```

```python
child_obj=child()
```

```python
child_obj.t_parent()
```

```python
class class1:
  def test_class1(self):
    print("this is my class1")
```

```python
class class2(class1):
  def test_class2(self):
    print("this is my class2")
```

```python
class class3(class2):
  def test_class3(self):
    print("this is my class3")
```

```python
obj_class3=class3()
```

```python
obj_class3.test_class1()
```

```python
obj_class3.test_class2()
```
