```python
l=[1,2,3,4,4,"sudh","kumar",324,34.678,"abc"]
```

```python
l1=[]
l2=[]
for i in l:
  if type(i)==int or type(i)== float:
    l1.append(i)
  else:
    l2.append(i)
print(l1)
print(l2)
```

```python
l1=[]
l2=[]
l3=[]
for i in l:
  if type(i)==int:
    l1.append(i)
  elif type(i)== float:
    l2.append(i)
  elif type(i)== str:
    l3.append(i)
print(l1)
print(l2)
print(l3)
```

```python
n=int(input("number"))
m=n
rev=0
while m>0:
 d=m%10
 rev=(rev*10)+d
 m=m//10
if rev==n:
 print("is palindrome")
else:
 print("not")
```
