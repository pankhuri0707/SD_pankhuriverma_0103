```python
def wid():
  amount = int(input("Enter Amount: "))


  if amount <= 0:
    print("Invalid Amount")

  elif amount > accounts[card]["balance"]:
    print("Insufficient Funds")
    print("Balance:", accounts[card]["balance"])

  else:
    atm_cash = 50000

  if amount > atm_cash:
    print("ATM Out of Cash")

  else:
    accounts[card]["balance"] -= amount
    print("Please Collect Your Cash")
    print("Updated Balance:", accounts[card]["balance"])
def dep():
  amount = int(input("Enter Deposit Amount: "))

  if amount <= 0:
    print("Invalid Amount")

  else:
    accounts[card]["balance"] += amount
    print("Deposit Successful")
    print("Updated Balance:", accounts[card]["balance"])
def bal():
  print("Current Balance:", accounts[card]["balance"])


accounts = {
    "1234567890": {"pin": "1234", "balance": 5000},
    "9876543210": {"pin": "4321", "balance": 10000}
}

card = input("Enter Card Number: ")

if card in accounts:
    pin = input("Enter PIN: ")

    if pin == accounts[card]["pin"]:

        while True:
            print("\n----- ATM MENU -----")
            print("1. Withdraw")
            print("2. Deposit")
            print("3. Balance Inquiry")
            print("4. Exit")

            choice = input("Enter your choice: ")
            if choice == "1":
                wid()
            elif choice == "2":
                dep()
            elif choice == "3":
                bal()
            elif choice == "4":
                print("Card Ejected")
                print("Thank You!")
                break

            else:
                print("Invalid Choice")

            receipt = input("Print Receipt? (y/n): ")
            if receipt.lower() == "y":
                print("----- RECEIPT -----")
                print("Balance:", accounts[card]["balance"])

            again = input("Another Transaction? (y/n): ")
            if again.lower() != "y":
                print("Card Ejected")
                print("Thank You!")
                break

    else:
        print("Incorrect PIN")
        print("Card Ejected")

else:
    print("Invalid Card Number")
```
